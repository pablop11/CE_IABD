2. Quédate solo con las columnas "asin" (el código de producto) y "overall" (la nota que le dan en la opinión).
3. Crea un nuevo dataframe que muestre cada uno de los productos (su asin) con la puntuación media que ha obtenido en las opiniones y el número de opiniones que tiene.

```python
import pandas as pd

df_instrumentos = pd.read_json("Musical_Instruments_5.json", lines=True)
df_instrumentos.head()
```

```python


df_instrumentos_filtrados = df_instrumentos[["asin", "overall"]]
df_instrumentos_filtrados.head()
```

```python
import numpy as np
import pandas as pd

num_asin = []
medias = []
num_overalls = []
modas = []

'''
df_productos = df_instrumentos_filtrados.groupby("asin").agg(
    puntuacion_media=("overall", "mean"),
    numero_opiniones=("overall", "count")
).reset_index()

df_productos
'''

for index, row in df_instrumentos_filtrados.iterrows():
    # comprobar si asin esta en num_asin
    # si no esta, lo añade y si esta no
    if row["asin"] not in num_asin:
        num_asin.append(row["asin"])

        # filtrar el dataframe para coger todos los overall que coinciden con el asin
        df_filtrado = df_instrumentos_filtrados.loc[df_instrumentos_filtrados["asin"] == row["asin"]]

        # meto la media de esos overall en medias
        medias.append(df_filtrado["overall"].mean())

        # meto la longitud de la lista de overalls en num_overalls
        num_overalls.append(len(df_filtrado))

        # calcular la moda de los overhall
        if not df_filtrado["overall"].mode().empty:
            modas.append(df_filtrado["overall"].mode().iloc[0])
        else:
            # si no hay valores ponemos nan
            modas.append(np.nan)  

df_resultado = pd.DataFrame({
    "asin": num_asin,
    "media_overall": medias,
    "num_overalls": num_overalls,
    "moda": modas
})
df_resultado.head()
```

```python
def elegir_producto(df, reviewer_id):
    # filtrar el dataframe por reviewerID
    fila_reviewer = df.loc[df["reviewerID"] == reviewer_id]
    
    # comprobar si se encontro alguna fila
    if not fila_reviewer.empty:
        # devolver el asin de la primera coincidencia
        return fila_reviewer.iloc[0]["asin"]
    else:
        return "Producto no encontrado"  

asin_elegido = elegir_producto(df_instrumentos, "A14VAT5EAX3D9S")
print("ASIN del producto elegido:", asin_elegido)
```
